```python
# json.dumps() 是将 python 对象转化为 json
# json.loads() 是将 json 转化为 python 对象

dict_data = {"a": 1, "b": 2}
print(dict_data)
print(type(dict_data),dict_data)
```

```python
# 将dict格式数据转换成json格式字符串
dump_data = json.dumps(dict_data)
print(dump_data)
print(type(dump_data),dump_data)
```

```python
# 将json格式字符串转换成对应的python值
load_data = json.loads(dump_data)
print(load_data)
print(type(load_data),load_data)
```

```python
import requests
import urllib.request
net163music = 'http://music.163.com/api/v1/resource/comments/R_SO_4_483671599?limit=10&offset=0'
headers = {'User-Agent':'chrome'}
response = requests.get(net163music, headers=headers,timeout=5)
response
```

```python
musicdata = response.json()
musicdata
```

```python
musicdata['hotComments'][8]['content']
```

```python
musicdata['hotComments'][8]['likedCount']
```

```python
musicdatafliter = []
musicdatahot = musicdata['hotComments']
musicdatahot
```

```python
import jsonpath
for job in musicdatahot:
    dic={}
    dic['content']=jsonpath.jsonpath(job,'$..content')[0] #评论
    dic['userId']=jsonpath.jsonpath(job['user'],'$..userId')[0]  #用户ID
    dic['nickname']=jsonpath.jsonpath(job['user'],'$..nickname')[0]#用户名
    dic['likedCount']=jsonpath.jsonpath(job,'$..likedCount')[0] #赞数
    musicdatafliter.append(dic)
musicdatafliter = pd.DataFrame(musicdatafliter)
musicdatafliter
```
