```python
import requests
import urllib.request
net163music = 'http://music.163.com/api/v1/resource/comments/R_SO_4_483671599?limit=10&offset=0'
headers = {'User-Agent':'chrome'}
response = requests.get(net163music, headers=headers,timeout=5)
response
```

```python
musicdata = response.json()
musicdata
```

```python
musicdatafliter = []
musicdatahot = musicdata['hotComments']
musicdatahot
```

```python
import jsonpath
for job in musicdatahot:
    dic={}
    dic['content']=jsonpath.jsonpath(job,'$..content')[0] #è¯„è®º
    dic['userId']=jsonpath.jsonpath(job['user'],'$..userId')[0]  #ç”¨æˆ·ID
    dic['nickname']=jsonpath.jsonpath(job['user'],'$..nickname')[0]#ç”¨æˆ·å
    dic['likedCount']=jsonpath.jsonpath(job,'$..likedCount')[0] #èµæ•°
    musicdatafliter.append(dic)
musicdatafliter = pd.DataFrame(musicdatafliter)
musicdatafliter
```
